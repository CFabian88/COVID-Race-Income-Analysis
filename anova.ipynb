```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
os.chdir('/Users/fabian_coll/Desktop/PyProjects/COVID-Race-Income-Analysis/data/clean')
```

```python
df = pd.read_csv('COVID_deaths.csv')
df = df.drop(['Unnamed: 0'], axis = 1)
df.head()
```

```python
# Look at # of Missing data points per race
races = ['White', 'Black', 'AmericanIndian/NativeIslander', 'Asian', 'NativeHawaiian/PacificIslander', 'Hispanic', 'Other']
xlabels = ['White', 'Black', 'AIndian/NIslander', 'Asian', 'NHawaiian/PIslander', 'Hispanic', 'Other']
race_num_missingVals = []
for i in races:
    rowName = str(i+'_per100k')
    race_num_missingVals.append(df[rowName].isna().sum())

plt.bar(xlabels, race_num_missingVals)
plt.title('# Missing Data Points by Race (out of 723)')
plt.xlabel('Race')
plt.ylabel('# Data Points')
plt.xticks(rotation = 60)
plt.show()
```
